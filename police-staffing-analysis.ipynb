# Police Workload and Staffing Analysis

Here, we import the necessary libraries, retrieve the data, and validate the app token for fetching without throttling limits.

```python
import pandas as pd
from sodapy import Socrata

APP_TOKEN = "CZbPwiKRv9CZsXufFK8jWajSY"
client = Socrata("data.seattle.gov", APP_TOKEN)

# Socrata's WHERE clause to filter by year 2025
where_clause = "year(report_date_time) = 2025"

# Pull in chunks of 1000 until no more results
all_results = []
offset = 0
limit = 1000

while True:
    batch = client.get("tazs-3rd5", where=where_clause, limit=limit, offset=offset)
    if not batch:
        break
    all_results.extend(batch)
    offset += limit

# Convert to DataFrame
df = pd.DataFrame.from_records(all_results)

# Convert 'report_date_time' to datetime
df['report_date_time'] = pd.to_datetime(df['report_date_time'])

# Sort by date
df = df.sort_values('report_date_time').reset_index(drop=True)

print(df.info())
print(df.head())
```
